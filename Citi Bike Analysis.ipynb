# Dependencies

```python
import pandas as pd
import os
```

```python
jan2018_file = pd.read_csv('./data/JC-201801-citibike-tripdata.csv')
jan2018_df = pd.DataFrame(jan2018_file)

feb2018_file = pd.read_csv('./data/JC-201802-citibike-tripdata.csv')
feb2018_df = pd.DataFrame(feb2018_file)

mar2018_file = pd.read_csv('./data/JC-201803-citibike-tripdata.csv')
mar2018_df = pd.DataFrame(mar2018_file)

apr2018_file = pd.read_csv('./data/JC-201804-citibike-tripdata.csv')
apr2018_df = pd.DataFrame(apr2018_file)

may2018_file = pd.read_csv('./data/JC-201805-citibike-tripdata.csv')
may2018_df = pd.DataFrame(may2018_file)

jun2018_file = pd.read_csv('./data/JC-201806-citibike-tripdata.csv')
jun2018_df = pd.DataFrame(jun2018_file)

jul2018_file = pd.read_csv('./data/JC-201807-citibike-tripdata.csv')
jul2018_df = pd.DataFrame(jul2018_file)

aug2018_file = pd.read_csv('./data/JC-201808-citibike-tripdata.csv')
aug2018_df = pd.DataFrame(aug2018_file)

sep2018_file = pd.read_csv('./data/JC-201809-citibike-tripdata.csv')
sep2018_df = pd.DataFrame(sep2018_file)

oct2018_file = pd.read_csv('./data/JC-201810-citibike-tripdata.csv')
oct2018_df = pd.DataFrame(oct2018_file)

nov2018_file = pd.read_csv('./data/JC-201811-citibike-tripdata.csv')
nov2018_df = pd.DataFrame(nov2018_file)

dec2018_file = pd.read_csv('./data/JC-201812-citibike-tripdata.csv')
dec2018_df = pd.DataFrame(dec2018_file)

dec2018_df.head()
```

```python
jan2018_df["month"] = "1/31/2018"
feb2018_df["month"] = "2/28/2018"
mar2018_df["month"] = "3/31/2018"
apr2018_df["month"] = "4/30/2018"
may2018_df["month"] = "5/31/2018"
jun2018_df["month"] = "6/30/2018"
jul2018_df["month"] = "7/31/2018"
aug2018_df["month"] = "8/31/2018"
sep2018_df["month"] = "9/30/2018"
oct2018_df["month"] = "10/31/2018"
nov2018_df["month"] = "11/30/2018"
dec2018_df["month"] = "12/31/2018"

dec2018_df.head()
```

```python
citibike2018_df = pd.concat([jan2018_df, feb2018_df, mar2018_df, apr2018_df, may2018_df, jun2018_df, jul2018_df, aug2018_df, sep2018_df, oct2018_df, nov2018_df, dec2018_df])
citibike2018_df.head()
# citibike2018_df["start station id"].count()
```

```python
citibike2018_df = citibike2018_df.rename(columns= {"starttime" : "startdatetime", "stoptime" : "stopdatetime"})

startdatetimesplit = citibike2018_df["startdatetime"].str.split(" ", expand=True)
citibike2018_df["startdate"] = startdatetimesplit[0]
citibike2018_df["starttime"] = startdatetimesplit[1]

stopdatetimesplit = citibike2018_df["stopdatetime"].str.split(" ", expand=True)
citibike2018_df["stopdate"] = stopdatetimesplit[0]
citibike2018_df["stoptime"] = stopdatetimesplit[1]
```

```python
# Clean Gender column
citibike2018_df["gender"] = citibike2018_df["gender"].replace(1, "male")
citibike2018_df["gender"] = citibike2018_df["gender"].replace(2, "female")
citibike2018_df["gender"] = citibike2018_df["gender"].replace(0, "unknown")

citibike2018_df.head()
```

```python
citibike2018_df.to_csv("./data/citibike2018_data.csv")
```
